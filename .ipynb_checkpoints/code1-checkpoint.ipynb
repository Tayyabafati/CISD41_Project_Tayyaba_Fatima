## Cleaning The Data

```python
#changing the datatype of Country/Territory and code from object to string
deathdf[['Country/Territory','Code']] = deathdf[['Country/Territory','Code']].astype('string')
#show the datatypes after changing.
deathdf.dtypes
```

```python
#chceking the unique values of the Country/Territory column
deathdf['Country/Territory'].unique()
```

```python
#checking the number of countries
len(deathdf['Country/Territory'].unique()) 
```

```python
#checking the unique values of Year Column.
deathdf['Year'].unique()
```

```python
#checking the no of years
len(deathdf['Year'].unique())
```

```python
#Renaming the column name 'Country/Territory' with ENTITY
deathdf = deathdf.rename(columns={'Country/Territory': 'Entity'})
```

The data is for 204 countries for about 30 years.

As we know there are number of diseases which falls into same category as nutritional deficiency
and protein-energy malnutritional falls into nutritional diseases so I merge the columns and 
make this data more organise because this dataset have 32 columns and so I make it to 22 columns.

```python
deathdf['Brain Disease'] = deathdf["Meningitis"]+deathdf["Alzheimer's Disease and Other Dementias"]+deathdf["Parkinson's Disease"]+deathdf["Neonatal Disorders"]                                               
```

```python
deathdf['Nutritional Disease'] = deathdf["Protein-Energy Malnutrition"]+deathdf['Nutritional Deficiencies']
```

```python
deathdf["Accidental"] = deathdf["Road Injuries"]+deathdf["Fire, Heat, and Hot Substances"] +deathdf["Drowning"]
```

```python
deathdf["Liver Diseases"] = deathdf["Cirrhosis and Other Chronic Liver Diseases"]+deathdf["Acute Hepatitis"]
```

```python
deathdf["Violence"] = deathdf["Interpersonal Violence"]+deathdf["Conflict and Terrorism"]
```

```python
deathdf["Respiratory Diseases"] = deathdf["Lower Respiratory Infections"]+deathdf["Chronic Respiratory Diseases"]
```

```python
deathdf["Infectious Diseases"] = deathdf["Malaria"]+deathdf["Diarrheal Diseases"]
```

```python
deathdf["Climatic"] = deathdf["Exposure to Forces of Nature"]+deathdf["Environmental Heat and Cold Exposure"]
```

```python
deathdf["Depressants"] = deathdf["Drug Use Disorders"]+deathdf["Alcohol Use Disorders"]
```

```python
#after merging the columns into one new column I am now dropping the column which I already merge.
dfnew = deathdf.drop(["Meningitis","Alzheimer's Disease and Other Dementias","Parkinson's Disease","Neonatal Disorders",
             "Protein-Energy Malnutrition","Nutritional Deficiencies","Drowning","Road Injuries",
             "Fire, Heat, and Hot Substances","Cirrhosis and Other Chronic Liver Diseases",
             "Acute Hepatitis","Interpersonal Violence","Conflict and Terrorism",
              "Chronic Respiratory Diseases","Lower Respiratory Infections","Malaria",
             "Diarrheal Diseases","Exposure to Forces of Nature","Environmental Heat and Cold Exposure"
             ,"Alcohol Use Disorders","Drug Use Disorders"],1)
```

```python
dfnew.head()
```

```python
#computing the total no of deaths per year by dropping the 'YEAR' column and putting it into a new column.
dfnew['Total_Deaths'] = dfnew.drop('Year', axis=1).sum(axis=1)
dfnew.head() #showing the head of the new dataframe with total
```

Now let's see the data for neighbouring countries for top 3 countries.
Let's implement a function by which we can easily see **top 10 diseases** data for **any country** by plotting it

```python
def top_3_illnesses(country):
    dfnew_country = dfnew[dfnew['Entity'] == country]
    min_year,max_year = dfnew_country['Year'].min(),dfnew_country['Year'].max()
    fig = px.scatter(dfnew_country.iloc[:, 5:21].sum().sort_values(ascending=False).head(4), title='Top 10 Cause of Deaths in {} between {} and {}'.format(country, min_year, max_year), color_discrete_sequence=px.colors.qualitative.Pastel1, labels={'value': 'Amount of deaths', 'index': 'Cause of Deaths'})
    fig.update_layout(showlegend=False)
    fig.show()
```

```python
#plotting Russia neighbouring country of China
top_3_illnesses('Russia')
```

```python
#plotting sri Lanka neighbouring country of India
top_3_illnesses('Sri Lanka')
```

In Sri Lanka the top 3 diseases are in top 3.

```python
#plotting data for pakistan which is the neighbouring country of China and India
top_3_illnesses('Pakistan')
```

```python
#getting the diseases deaths data of Mexico which is the neigbouring country of United States
top_3_illnesses('Mexico')
```
